### 3. Extracci√≥n de las Cartas

Esta celda define y ejecuta la funci√≥n principal para obtener el texto de entrada.

#### 3.1 Definici√≥n de `obtener_cartas_frankenstein()`

Esta funci√≥n es la responsable de conectarse a la web del Proyecto Gutenberg, descargar el texto completo de "Frankenstein" y extraer limpiamente las cuatro cartas introductorias.

**Proceso detallado:**

1.  **Descarga:** Utiliza `requests.get()` para descargar el contenido de la URL de `pg84.txt`.
2.  **Pre-procesamiento:**
    * Normaliza todos los saltos de l√≠nea a `\n` para consistencia.
    * **Salto de √çndice (Clave):** Localiza el marcador `*** START OF` que indica el inicio del contenido real y avanza hasta el siguiente p√°rrafo (doble salto de l√≠nea) para omitir el √≠ndice y la metadata de Project Gutenberg.
3.  **Extracci√≥n con Expresiones Regulares (Regex):**
    * Se usa `re.search()` para cada carta con patrones espec√≠ficos.
    * **Cartas 1-3:** El patr√≥n busca `(Letter X ...)` y captura todo (`[\s\S]+?`) hasta que "ve" el inicio de la siguiente carta (`(?=Letter X+1...)`). Se usa un *lookahead* (`?=`) para que el texto de la siguiente carta no sea consumido y pueda ser encontrado por la siguiente b√∫squeda.
    * **Carta 4:** El patr√≥n es similar, pero su delimitador final es el inicio del `Chapter 1` o `CHAPTER I`.
4.  **Verificaci√≥n:** La funci√≥n imprime el estado de la extracci√≥n (si se encontr√≥ o no cada carta) y la longitud del texto extra√≠do.
5.  **Retorno:** Devuelve una lista (`cartas`) donde cada elemento es un *string* con el texto de una carta.

#### 3.2 Ejecuci√≥n y Validaci√≥n

La segunda parte de la celda ejecuta la funci√≥n `obtener_cartas_frankenstein()` y almacena el resultado en la variable `cartas`.

1.  **Ejecuci√≥n:** Se llama a la funci√≥n y se muestra el *log* de extracci√≥n (ej. `‚úì Carta 1 extra√≠da...`).
2.  **Resultados:** Imprime un resumen de cada carta extra√≠da, mostrando:
    * Longitud total en caracteres.
    * Conteo de palabras (aproximado).
    * Los primeros 250 y los √∫ltimos 150 caracteres como vista previa.
3.  **Validaci√≥n:** Compara la longitud de cada carta extra√≠da con un rango de longitudes esperadas (`rangos`). Esto sirve como un "sanity check" para asegurar que la extracci√≥n fue exitosa y no se cort√≥ el texto o se extrajo texto de m√°s (p.ej., si el formato de la web de Gutenberg cambiara).

```python
def obtener_cartas_frankenstein():
    """
    Descarga y extrae las 4 primeras cartas del libro Frankenstein de Project Gutenberg.
    
    Returns:
        list: Lista con las 4 cartas extra√≠das como strings
    """
    url = "https://www.gutenberg.org/cache/epub/84/pg84.txt"
    response = requests.get(url)
    texto = response.text
    
    # Normalizar saltos de l√≠nea
    texto = texto.replace('\r\n', '\n').replace('\r', '\n')
    
    # CLAVE: Saltar el √≠ndice/tabla de contenidos
    # Buscar donde empieza realmente el contenido (despu√©s de "*** START OF")
    inicio_contenido = texto.find('*** START OF')
    if inicio_contenido != -1:
        # Buscar 500 caracteres despu√©s del marcador de inicio
        inicio_contenido = texto.find('\n\n', inicio_contenido + 100)
        texto = texto[inicio_contenido:]
        print(f"‚úì Saltado el √≠ndice (inicio en posici√≥n {inicio_contenido})")
    
    # Ahora buscar las cartas con un patr√≥n m√°s espec√≠fico
    # Buscar "Letter X" seguido de "_To Mrs. Saville"
    carta1 = re.search(
        r'(Letter 1\s+_To Mrs\. Saville, England\._[\s\S]+?)(?=Letter 2\s+_To Mrs)', 
        texto
    )
    carta2 = re.search(
        r'(Letter 2\s+_To Mrs\. Saville, England\._[\s\S]+?)(?=Letter 3\s+_To Mrs)', 
        texto
    )
    carta3 = re.search(
        r'(Letter 3\s+_To Mrs\. Saville, England\._[\s\S]+?)(?=Letter 4\s+_To Mrs)', 
        texto
    )
    carta4 = re.search(
        r'(Letter 4\s+_To Mrs\. Saville, England\._[\s\S]+?)(?=\n\s*Chapter 1\s+\n|\n\s*CHAPTER I\s+\n)', 
        texto
    )
    
    cartas = []
    
    if carta1:
        cartas.append(carta1.group(1).strip())
        print(f"‚úì Carta 1 extra√≠da: {len(carta1.group(1)):,} caracteres")
    else:
        print("‚úó Carta 1 NO encontrada")
    
    if carta2:
        cartas.append(carta2.group(1).strip())
        print(f"‚úì Carta 2 extra√≠da: {len(carta2.group(1)):,} caracteres")
    else:
        print("‚úó Carta 2 NO encontrada")
    
    if carta3:
        cartas.append(carta3.group(1).strip())
        print(f"‚úì Carta 3 extra√≠da: {len(carta3.group(1)):,} caracteres")
    else:
        print("‚úó Carta 3 NO encontrada")
    
    if carta4:
        cartas.append(carta4.group(1).strip())
        print(f"‚úì Carta 4 extra√≠da: {len(carta4.group(1)):,} caracteres")
    else:
        print("‚úó Carta 4 NO encontrada")
    
    # Verificaci√≥n
    if len(cartas) == 4:
        print(f"\n‚úÖ Se extrajeron las 4 cartas correctamente")
    else:
        print(f"\n‚ö†Ô∏è Se extrajeron {len(cartas)} cartas en lugar de 4")
    
    return cartas

# Ejecutar extracci√≥n
print("=" * 80)
print("EXTRACCI√ìN DE CARTAS DE FRANKENSTEIN")
print("=" * 80 + "\n")

cartas = obtener_cartas_frankenstein()

# Mostrar resultados
print("\n" + "=" * 80)
print(f"RESULTADO FINAL: {len(cartas)} cartas extra√≠das")
print("=" * 80)

for i, carta in enumerate(cartas):
    print(f"\nüìß CARTA {i+1}")
    print(f"   Longitud: {len(carta):,} caracteres")
    print(f"   Palabras: {len(carta.split()):,}")
    print(f"   Primeras 250 caracteres:")
    print(f"   {carta[:250]}")
    print(f"   √öltimas 150 caracteres:")
    print(f"   ...{carta[-150:]}")
    print("-" * 80)

# Validaci√≥n
print("\nüìä VALIDACI√ìN:")
rangos = [(6000, 8000), (7000, 9000), (1500, 2500), (8000, 15000)]
for i, carta in enumerate(cartas):
    if i < len(rangos):
        min_esp, max_esp = rangos[i]
        longitud = len(carta)
        status = "‚úÖ" if min_esp <= longitud <= max_esp else "‚ö†Ô∏è"
        print(f"{status} Carta {i+1}: {longitud:,} caracteres (esperado: {min_esp:,}-{max_esp:,})")
```

## 5. Normalizaci√≥n de Textos (Justificaci√≥n)

[cite_start]Esta celda define las funciones de normalizaci√≥n, cumpliendo con el **Punto 2 de la pr√°ctica**.

El requisito clave es **justificar** por qu√© se normaliza el texto de manera diferente para cada algoritmo. No todos los algoritmos procesan el texto de la misma forma; aplicar una normalizaci√≥n √∫nica y agresiva (como quitar toda la puntuaci√≥n o convertir todo a min√∫sculas) puede ser beneficioso para algunos, pero perjudicial para otros.

A continuaci√≥n, se define una funci√≥n de normalizaci√≥n espec√≠fica para cada uno de los 6 algoritmos.

#### 5.1 Justificaci√≥n de las T√©cnicas

Las diferencias en las funciones se basan en c√≥mo funciona cada algoritmo:

1.  **Para TF-IDF, Frecuencia, TextRank y LSA:**
    * Estos m√©todos se basan en estad√≠sticas de palabras o en la estructura de las oraciones.
    * **Acci√≥n:** Se eliminan caracteres que no sean alfanum√©ricos, **excepto la puntuaci√≥n b√°sica (., !, ?)**.
    * **Justificaci√≥n:** Se conserva esta puntuaci√≥n esencial para que el tokenizador de oraciones (`sent_tokenize` de NLTK) pueda dividir el texto en oraciones correctamente. La mayor√≠a de los dem√°s s√≠mbolos (`;`, `*`, `(`, `)`) se eliminan por ser "ruido" para estos modelos.
    * **LSA (Extra):** La funci√≥n `normalizar_para_lsa` adem√°s convierte todo a min√∫sculas (`.lower()`). Esto es vital para los modelos de "bolsa de palabras" (Bag-of-Words) como LSA y TF-IDF, para que "Pole" y "pole" se traten como la misma palabra. (Nota: `TfidfVectorizer` y `CountVectorizer` a menudo hacen esto internamente, pero aqu√≠ se hace expl√≠cito para LSA).

2.  **Para RAKE (`normalizar_para_rake`):**
    * RAKE (Rapid Automatic Keyword Extraction) funciona identificando "frases" que est√°n delimitadas por palabras vac√≠as (*stopwords*) **y por puntuaci√≥n**.
    * **Acci√≥n:** **No se elimina casi ninguna puntuaci√≥n**. Solo se eliminan los saltos de l√≠nea y los subt√≠tulos (`_..._`).
    * **Justificaci√≥n:** Si elimin√°ramos la puntuaci√≥n (como comas o puntos y coma), el algoritmo RAKE fusionar√≠a frases clave y perder√≠a su capacidad de identificarlas correctamente.

3.  **Para BERT (`normalizar_para_bert`):**
    * BERT es un modelo contextual profundo. Fue entrenado con texto "crudo" de Internet y entiende el significado de la puntuaci√≥n, las may√∫sculas y el contexto.
    * **Acci√≥n:** Es la normalizaci√≥n **m√°s ligera**. Solo se eliminan los saltos de l√≠nea, los t√≠tulos "Letter X" y los espacios extra.
    * **Justificaci√≥n:** Eliminar puntuaci√≥n o texto (como los subt√≠tulos `_To Mrs. Saville_`) podr√≠a *empeorar* los resultados, ya que le estar√≠amos quitando al modelo informaci√≥n contextual que sabe c√≥mo interpretar para entender el significado sem√°ntico de la oraci√≥n.

```python
def normalizar_para_tfidf(texto):
    texto = texto.replace('\r\n', ' ')
    texto = texto.replace('Letter 1', '').replace('Letter 2', '').replace('Letter 3', '').replace('Letter 4', '')
    texto = re.sub(r'_.*?_', ' ', texto)
    texto = re.sub(r'[^a-zA-Z0-9\s.,!?]', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

def normalizar_para_frecuencia(texto):
    texto = texto.replace('\r\n', ' ')
    texto = texto.replace('Letter 1', '').replace('Letter 2', '').replace('Letter 3', '').replace('Letter 4', '')
    texto = re.sub(r'_.*?_', ' ', texto)
    texto = re.sub(r'[^a-zA-Z0-9\s.,!?]', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

def normalizar_para_rake(texto):
    texto = texto.replace('\r\n', ' ')
    texto = texto.replace('Letter 1', '').replace('Letter 2', '').replace('Letter 3', '').replace('Letter 4', '')
    texto = re.sub(r'_.*?_', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

def normalizar_para_textrank(texto):
    texto = texto.replace('\r\n', ' ')
    texto = texto.replace('Letter 1', '').replace('Letter 2', '').replace('Letter 3', '').replace('Letter 4', '')
    texto = re.sub(r'_.*?_', ' ', texto)
    texto = re.sub(r'[^a-zA-Z0-9\s.,!?]', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

def normalizar_para_bert(texto):
    texto = texto.replace('\r\n', ' ')
    texto = texto.replace('Letter 1', '').replace('Letter 2', '').replace('Letter 3', '').replace('Letter 4', '')
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

def normalizar_para_lsa(texto):
    texto = texto.replace('\r\n', ' ')
    texto = texto.replace('Letter 1', '').replace('Letter 2', '').replace('Letter 3', '').replace('Letter 4', '')
    texto = re.sub(r'_.*?_', ' ', texto)
    texto = re.sub(r'[^a-zA-Z0-9\s.,!?]', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    texto = texto.lower()
    return texto
```

### 7. Definici√≥n de Funciones de Resumen (Punto 3)

Esta celda define las 6 funciones de resumen autom√°tico, cumpliendo con el **Punto 3 de la pr√°ctica**El objetivo es generar un resumen extrayendo las **12 frases m√°s representativas** de cada carta.

Todas las funciones son **extractivas**: no generan texto nuevo, sino que implementan una l√≥gica diferente para asignar un "puntaje" a cada oraci√≥n original. Luego, seleccionan las 12 oraciones con el puntaje m√°s alto.

Una caracter√≠stica com√∫n e importante es que, despu√©s de encontrar las 12 mejores oraciones, sus √≠ndices se reordenan (`indices_top = sorted(indices_top)`). Esto es crucial para que el resumen final se lea de forma coherente y en el orden cronol√≥gico original.

#### 7.1 `resumen_tfidf` (TF-IDF) 
* **L√≥gica:** Trata cada oraci√≥n como un "documento" y cada palabra como un "t√©rmino".
* **Puntaje:** El puntaje de una oraci√≥n es la **suma de los puntajes TF-IDF de todas las palabras** que contiene.
* **Intuici√≥n:** Una oraci√≥n es importante si contiene palabras que son frecuentes en *esa* oraci√≥n (TF) pero raras en *todas las dem√°s oraciones* del texto (IDF).

#### 7.2 `resumen_frecuencia` (Frecuencia Normalizada) 
* **L√≥gica:** Se basa en la idea de que las oraciones que contienen las palabras m√°s frecuentes (excluyendo *stopwords*) son las m√°s importantes.
* **Puntaje:**
    1.  Calcula la frecuencia de cada palabra en el texto y la normaliza (dividi√©ndola por la frecuencia de la palabra m√°s com√∫n).
    2.  El puntaje de una oraci√≥n es el **promedio de los puntajes de frecuencia normalizada** de las palabras que la componen.

#### 7.3 `resumen_rake` (RAKE) 
* **L√≥gica:** RAKE (Rapid Automatic Keyword Extraction) es un algoritmo que extrae frases clave bas√°ndose en delimitadores (como *stopwords* y puntuaci√≥n).
* **Puntaje:**
    1.  Primero, RAKE extrae una lista de frases clave y sus puntajes de todo el texto.
    2.  El puntaje de una oraci√≥n es la **suma de los puntajes de las frases clave (RAKE) que aparecen en ella**.
* **Intuici√≥n:** Una oraci√≥n es importante si contiene muchas de las frases clave m√°s relevantes del documento.

#### 7.4 `resumen_textrank` (TextRank)
* **L√≥gica:** Aplica el algoritmo PageRank de Google al texto. 
* **Puntaje:**
    1.  Crea un **grafo** donde cada oraci√≥n es un nodo.
    2.  La arista (conexi√≥n) entre dos nodos se pondera seg√∫n su **similitud** (en este c√≥digo, se usa la similitud TF-IDF).
    3.  Se ejecuta el algoritmo `nx.pagerank()` en el grafo.
* **Intuici√≥n:** Una oraci√≥n es importante si **muchas otras oraciones (que tambi√©n son importantes) son similares a ella**.

#### 7.5 `resumen_bert` (BERT)
* **L√≥gica:** Utiliza *embeddings* contextuales profundos para entender el **significado sem√°ntico** del texto.
* **Optimizaci√≥n Clave:** Esta funci√≥n recibe el `MODELO_BERT` pre-cargado como par√°metro. Esto es vital para el rendimiento, ya que evita que el pesado modelo se cargue desde cero en cada llamada.
* **Puntaje:**
    1.  Genera un vector (embedding) para el **documento completo** y un vector para **cada oraci√≥n individual**.
    2.  El puntaje de cada oraci√≥n es su **similitud coseno** con el vector del documento completo.
* **Intuici√≥n:** Una oraci√≥n es importante si su significado es **sem√°nticamente muy similar al significado del documento en su totalidad**.

#### 7.6 `resumen_lsa` (LSA)
* **L√≥gica:** Latent Semantic Analysis (An√°lisis Sem√°ntico Latente). Utiliza √°lgebra lineal (SVD) para encontrar los "t√≥picos" o "conceptos" latentes en el texto. 
* **Puntaje:**
    1.  Crea una matriz oraci√≥n-palabra (`CountVectorizer`).
    2.  Aplica **Descomposici√≥n de Valor Singular (SVD)** (`TruncatedSVD`) para reducir la matriz a sus componentes (t√≥picos) principales.
    3.  El puntaje de una oraci√≥n es la **suma de la magnitud (valor absoluto) de sus componentes** en esta matriz reducida.
* **Intuici√≥n:** Una oraci√≥n es importante si est√° fuertemente conectada a los **t√≥picos m√°s centrales** del documento.

```python
def resumen_tfidf(texto, num_oraciones=12):
    """Genera resumen usando TF-IDF con 12 oraciones"""
    oraciones = sent_tokenize(texto)
    
    if len(oraciones) < num_oraciones:
        num_oraciones = len(oraciones)
    
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(oraciones)
    
    puntajes = tfidf_matrix.sum(axis=1).A1
    
    indices_top = puntajes.argsort()[-num_oraciones:][::-1]
    indices_top = sorted(indices_top)
    
    resumen = [oraciones[i] for i in indices_top]
    return resumen


def resumen_frecuencia(texto, num_oraciones=12):
    """Genera resumen usando frecuencia normalizada con 12 oraciones"""
    oraciones = sent_tokenize(texto)
    
    if len(oraciones) < num_oraciones:
        num_oraciones = len(oraciones)
    
    palabras = word_tokenize(texto.lower())
    stop_words = set(stopwords.words('english'))
    palabras_filtradas = [palabra for palabra in palabras 
                          if palabra.isalpha() and palabra not in stop_words]
    
    frecuencia = {}
    for palabra in palabras_filtradas:
        frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
    
    if not frecuencia:
        return oraciones[:num_oraciones]
    
    max_freq = max(frecuencia.values())
    for palabra in frecuencia:
        frecuencia[palabra] = frecuencia[palabra] / max_freq
    
    puntajes = []
    for oracion in oraciones:
        palabras_oracion = word_tokenize(oracion.lower())
        puntaje = sum(frecuencia.get(palabra, 0) for palabra in palabras_oracion 
                     if palabra in frecuencia)
        contador = sum(1 for palabra in palabras_oracion if palabra in frecuencia)
        puntajes.append(puntaje / contador if contador > 0 else 0)
    
    indices_top = sorted(range(len(puntajes)), key=lambda i: puntajes[i], 
                        reverse=True)[:num_oraciones]
    indices_top = sorted(indices_top)
    
    resumen = [oraciones[i] for i in indices_top]
    return resumen


def resumen_rake(texto, num_oraciones=12):
    """Genera resumen usando RAKE con 12 oraciones"""
    oraciones = sent_tokenize(texto)
    
    if len(oraciones) < num_oraciones:
        num_oraciones = len(oraciones)
    
    r = Rake()
    r.extract_keywords_from_text(texto)
    frases = r.get_ranked_phrases_with_scores()
    
    puntajes = []
    for oracion in oraciones:
        puntaje = sum(puntuacion for puntuacion, frase in frases 
                     if frase.lower() in oracion.lower())
        puntajes.append(puntaje)
    
    indices_top = sorted(range(len(puntajes)), key=lambda i: puntajes[i], 
                        reverse=True)[:num_oraciones]
    indices_top = sorted(indices_top)
    
    resumen = [oraciones[i] for i in indices_top]
    return resumen


def resumen_textrank(texto, num_oraciones=12):
    """Genera resumen usando TextRank con 12 oraciones"""
    oraciones = sent_tokenize(texto)
    
    if len(oraciones) < num_oraciones:
        num_oraciones = len(oraciones)
    
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(oraciones)
    similarity_matrix = (tfidf_matrix * tfidf_matrix.T).toarray()
    
    graph = nx.from_numpy_array(similarity_matrix)
    scores = nx.pagerank(graph)
    
    top_indices = sorted(scores, key=scores.get, reverse=True)[:num_oraciones]
    top_indices = sorted(top_indices)
    resumen = [oraciones[i] for i in top_indices]
    return resumen

def resumen_bert(texto, num_oraciones=12, model=MODELO_BERT):
    """
    Genera resumen usando BERT con 12 oraciones
    
    MEJORA CR√çTICA: Recibe el modelo como par√°metro para evitar recargarlo
    IMPACTO: Reduce tiempo de ~6.7s total a ~1.5s total (4.5x m√°s r√°pido)
    
    Args:
        texto: Texto a resumir
        num_oraciones: N√∫mero de oraciones del resumen
        model: Modelo BERT pre-cargado (por defecto usa MODELO_BERT global)
    """
    oraciones = sent_tokenize(texto)
    
    if len(oraciones) < num_oraciones:
        num_oraciones = len(oraciones)
    
    # CLAVE: Usa el modelo pre-cargado, NO lo recarga
    embeddings_oraciones = model.encode(oraciones)
    embedding_documento = model.encode([texto])[0]
    
    similitudes = cosine_similarity([embedding_documento], embeddings_oraciones)[0]
    
    indices_top = sorted(range(len(similitudes)), key=lambda i: similitudes[i], 
                        reverse=True)[:num_oraciones]
    indices_top = sorted(indices_top)
    resumen = [oraciones[i] for i in indices_top]
    return resumen


def resumen_lsa(texto, num_oraciones=12):
    """Genera resumen usando LSA con 12 oraciones"""
    oraciones = sent_tokenize(texto)
    
    if len(oraciones) < num_oraciones:
        num_oraciones = len(oraciones)
    
    vectorizer = CountVectorizer(stop_words='english')
    document_matrix = vectorizer.fit_transform(oraciones)
    
    n_components = min(num_oraciones, len(oraciones), document_matrix.shape[1])
    svd = TruncatedSVD(n_components=n_components)
    document_matrix_reduced = svd.fit_transform(document_matrix)
    
    puntajes = [sum(abs(document_matrix_reduced[i])) for i in range(len(oraciones))]
    
    indices_top = sorted(range(len(puntajes)), key=lambda i: puntajes[i], 
                        reverse=True)[:num_oraciones]
    indices_top = sorted(indices_top)
    resumen = [oraciones[i] for i in indices_top]
    return resumen

print("‚úÖ Funciones de resumen definidas (12 oraciones con optimizaciones)")

```

#### 9.1 Proceso Detallado

1.  **Inicializaci√≥n:** Se crean dos conjuntos de 6 listas vac√≠as:
    * `tiempos_...` (ej. `tiempos_tfidf`): Para almacenar los tiempos de ejecuci√≥n (en segundos) de cada algoritmo.
    * `resumenes_...` (ej. `resumenes_tfidf`): Para almacenar el resultado del resumen (que es una lista de 12 oraciones) de cada algoritmo.

2.  **Bucle Principal:** El c√≥digo itera 4 veces, una por cada carta (`for i in range(4)`).

3.  **Ejecuci√≥n y Medici√≥n:** Dentro del bucle, para cada carta, se ejecutan los 6 algoritmos secuencialmente:
    * Se registra el tiempo de inicio (`start = time.time()`).
    * Se llama a la funci√≥n de resumen (ej. `resumen_tfidf(cartas_tfidf[i])`), pas√°ndole el texto normalizado correspondiente.
    * Se calcula el tiempo transcurrido y se a√±ade a la lista `tiempos_...`.
    * El resumen generado se a√±ade a la lista `resumenes_...`.
    * Se imprime un mensaje de estado (ej. `‚úì TF-IDF: 12 oraciones en 0.015s`).

4.  **Optimizaci√≥n de BERT:** Se puede observar en la llamada a `resumen_bert(cartas_bert[i])` que se est√° utilizando la versi√≥n optimizada, la cual recibe el modelo `MODELO_BERT` pre-cargado, resultando en una ejecuci√≥n mucho m√°s r√°pida.

Al finalizar el bucle, la celda imprime la **comparaci√≥n de tiempos promedio**.

* Calcula el promedio de tiempo para cada algoritmo (la suma de los 4 tiempos de ejecuci√≥n dividida entre 4).
* Esta tabla de promedios es un resultado fundamental para el **Punto 4: An√°lisis y Conclusiones**, ya que permite comparar la eficiencia computacional de cada m√©todo.

```python
# ============================================================================
# SE EJECUTAN LOS ALGORITMOS PARA LAS 4 CARTAS
# ============================================================================

print("\n" + "="*80)
print("GENERANDO RES√öMENES CON 12 ORACIONES POR CARTA")
print("="*80 + "\n")

# Listas para almacenar todo
tiempos_tfidf = []
tiempos_frecuencia = []
tiempos_rake = []
tiempos_textrank = []
tiempos_bert = []
tiempos_lsa = []

resumenes_tfidf = []
resumenes_frecuencia = []
resumenes_rake = []
resumenes_textrank = []
resumenes_bert = []
resumenes_lsa = []

# Ejecutar para cada carta
for i in range(4):
    print(f"\nüìß Procesando Carta {i+1}...")
    
    # TF-IDF
    start = time.time()
    res_tfidf = resumen_tfidf(cartas_tfidf[i])
    tiempos_tfidf.append(time.time() - start)
    resumenes_tfidf.append(res_tfidf)
    print(f"  ‚úì TF-IDF: {len(res_tfidf)} oraciones en {tiempos_tfidf[-1]:.3f}s")
    
    # Frecuencia
    start = time.time()
    res_freq = resumen_frecuencia(cartas_frecuencia[i])
    tiempos_frecuencia.append(time.time() - start)
    resumenes_frecuencia.append(res_freq)
    print(f"  ‚úì Frecuencia: {len(res_freq)} oraciones en {tiempos_frecuencia[-1]:.3f}s")
    
    # RAKE
    start = time.time()
    res_rake = resumen_rake(cartas_rake[i])
    tiempos_rake.append(time.time() - start)
    resumenes_rake.append(res_rake)
    print(f"  ‚úì RAKE: {len(res_rake)} oraciones en {tiempos_rake[-1]:.3f}s")
    
    # TextRank
    start = time.time()
    res_textrank = resumen_textrank(cartas_textrank[i])
    tiempos_textrank.append(time.time() - start)
    resumenes_textrank.append(res_textrank)
    print(f"  ‚úì TextRank: {len(res_textrank)} oraciones en {tiempos_textrank[-1]:.3f}s")
    
    # BERT (AHORA OPTIMIZADO)
    start = time.time()
    res_bert = resumen_bert(cartas_bert[i])  # Usa modelo pre-cargado
    tiempos_bert.append(time.time() - start)
    resumenes_bert.append(res_bert)
    print(f"  ‚úì BERT: {len(res_bert)} oraciones en {tiempos_bert[-1]:.3f}s")
    
    # LSA
    start = time.time()
    res_lsa = resumen_lsa(cartas_lsa[i])
    tiempos_lsa.append(time.time() - start)
    resumenes_lsa.append(res_lsa)
    print(f"  ‚úì LSA: {len(res_lsa)} oraciones en {tiempos_lsa[-1]:.3f}s")

print("\n‚úÖ Todos los res√∫menes generados correctamente")

# Mostrar comparaci√≥n de tiempos
print("\n" + "="*80)
print("COMPARACI√ìN DE TIEMPOS PROMEDIO")
print("="*80)
print(f"TF-IDF:      {sum(tiempos_tfidf)/4:.3f}s promedio")
print(f"Frecuencia:  {sum(tiempos_frecuencia)/4:.3f}s promedio")
print(f"RAKE:        {sum(tiempos_rake)/4:.3f}s promedio")
print(f"TextRank:    {sum(tiempos_textrank)/4:.3f}s promedio")
print(f"BERT:        {sum(tiempos_bert)/4:.3f}s promedio")
print(f"LSA:         {sum(tiempos_lsa)/4:.3f}s promedio")

```

### 12. An√°lisis Cualitativo de Contenido 

Esta celda complementa el an√°lisis de *velocidad* con un an√°lisis del *contenido* de los res√∫menes. El objetivo es cuantificar las diferencias cualitativas entre los resultados de cada algoritmo, lo cual es fundamental para el **Punto 4 (An√°lisis y Conclusiones)**.

El an√°lisis se divide en dos partes:
1.  **An√°lisis de Caracter√≠sticas (Tabla 2):** Mide las propiedades del texto de los res√∫menes (longitud, puntuaci√≥n, etc.).
2.  **An√°lisis de Similitud (Tabla 3 y Figura 5):** Mide qu√© tan parecidas son las *elecciones* de oraciones entre los algoritmos.

#### 12.1 An√°lisis de Caracter√≠sticas (Tabla 2)

La funci√≥n `analizar_resumenes` se crea para extraer un "perfil" de cada algoritmo:

* **Longitud Promedio / Palabras Promedio:** Mide si un algoritmo prefiere oraciones largas y descriptivas (como RAKE) o cortas y concisas.
* **Preserva Formato:** Busca puntuaci√≥n especial (como `‚Äî` o `_`). Esto es una **verificaci√≥n clave de la normalizaci√≥n (Punto 2)**. Se espera que RAKE y BERT den `True`, mientras que los otros (que limpiaron esa puntuaci√≥n) den `False`.
* **Usa May√∫sculas:** Esta es otra **verificaci√≥n de la normalizaci√≥n**. Se espera que LSA d√© `False`, ya que su funci√≥n `normalizar_para_lsa` inclu√≠a un `.lower()`.
* **Oraciones Cortas:** Mide la cantidad de oraciones "basura" o de baja calidad (menos de 50 caracteres), como fechas (ej. `St. Petersburgh...`). Un n√∫mero alto aqu√≠ (como se espera en `Frecuencia`) indica un resumen de menor calidad.

Los resultados se compilan en la **Tabla 2**, que ofrece una visi√≥n general de las propiedades de cada resumen.

#### 12.2 An√°lisis de Similitud (Tabla 3 y Figura 5)

Esta es la parte m√°s importante del an√°lisis cualitativo. Mide el "acuerdo" entre los algoritmos.

* La funci√≥n `calcular_solapamiento` compara los res√∫menes de dos algoritmos (para las 4 cartas) y devuelve el **porcentaje promedio de oraciones que ambos seleccionaron en com√∫n**.
* **Tabla 3 (Matriz de Similitud):** Muestra estos porcentajes. Un valor alto (ej. 70%) significa que dos algoritmos "piensan" de manera muy similar.
* **Figura 5 (Mapa de Calor):** Es la visualizaci√≥n de la Tabla 3.
    * **Cuadros "calientes" (rojos/naranjas):** Indican alta similitud. Se espera ver esto entre los m√©todos estad√≠sticos (TF-IDF, LSA, TextRank).
    * **Cuadros "fr√≠os" (amarillos):** Indican baja similitud. Se espera que BERT y Frecuencia sean "islas fr√≠as", ya que sus l√≥gicas de selecci√≥n (sem√°ntica y frecuencia simple) son √∫nicas y diferentes de las dem√°s.

Estas m√©tricas proporcionan la evidencia objetiva para la conclusi√≥n final sobre qu√© algoritmo produce los res√∫menes de mejor calidad y m√°s √∫nicos.

```python
# =============================================================================
# AN√ÅLISIS CUALITATIVO
# =============================================================================

print("=" * 80)
print("AN√ÅLISIS CUALITATIVO DE LOS RES√öMENES GENERADOS")
print("=" * 80)

# Funci√≥n para analizar caracter√≠sticas de los res√∫menes
def analizar_resumenes(resumenes, nombre_algoritmo):
    """Analiza caracter√≠sticas de los res√∫menes generados"""
    stats = {
        'algoritmo': nombre_algoritmo,
        'longitud_promedio': [],
        'palabras_promedio': [],
        'tiene_mayusculas': [],
        'mantiene_puntuacion': [],
        'oraciones_cortas': []  # < 50 caracteres
    }
    
    for carta_resumenes in resumenes:
        longitudes = [len(oracion) for oracion in carta_resumenes]
        palabras = [len(oracion.split()) for oracion in carta_resumenes]
        
        stats['longitud_promedio'].append(np.mean(longitudes))
        stats['palabras_promedio'].append(np.mean(palabras))
        stats['tiene_mayusculas'].append(
            any(oracion[0].isupper() for oracion in carta_resumenes if oracion)
        )
        stats['mantiene_puntuacion'].append(
            any('‚Äî' in oracion or '_' in oracion for oracion in carta_resumenes)
        )
        stats['oraciones_cortas'].append(
            sum(1 for l in longitudes if l < 50)
        )
    
    return stats

# Analizar cada algoritmo
analisis = []
for nombre, resumenes in [
    ("TF-IDF", resumenes_tfidf),
    ("Frecuencia", resumenes_frecuencia),
    ("RAKE", resumenes_rake),
    ("TextRank", resumenes_textrank),
    ("BERT", resumenes_bert),
    ("LSA", resumenes_lsa)
]:
    stats = analizar_resumenes(resumenes, nombre)
    analisis.append(stats)

# Crear DataFrame de an√°lisis
df_analisis = pd.DataFrame({
    'Algoritmo': [a['algoritmo'] for a in analisis],
    'Long. Promedio (chars)': [np.mean(a['longitud_promedio']) for a in analisis],
    'Palabras Promedio': [np.mean(a['palabras_promedio']) for a in analisis],
    'Preserva Formato': [any(a['mantiene_puntuacion']) for a in analisis],
    'Usa May√∫sculas': [all(a['tiene_mayusculas']) for a in analisis],
    'Oraciones Cortas': [np.mean(a['oraciones_cortas']) for a in analisis]
})

print("\n" + "=" * 80)
print("TABLA 2: CARACTER√çSTICAS CUALITATIVAS DE LOS RES√öMENES")
print("=" * 80)
print(df_analisis.to_string(index=False))
print("=" * 80)

# An√°lisis de solapamiento entre algoritmos
print("\n" + "=" * 80)
print("AN√ÅLISIS DE SIMILITUD ENTRE ALGORITMOS")
print("=" * 80)

def calcular_solapamiento(res1, res2):
    """Calcula el porcentaje de oraciones en com√∫n entre dos res√∫menes"""
    total_comparaciones = 0
    oraciones_comunes = 0
    
    for i in range(len(res1)):
        set1 = set(res1[i])
        set2 = set(res2[i])
        oraciones_comunes += len(set1.intersection(set2))
        total_comparaciones += len(set1)
    
    return (oraciones_comunes / total_comparaciones * 100) if total_comparaciones > 0 else 0

# Matriz de similitud
algoritmos_nombres = ['TF-IDF', 'Frecuencia', 'RAKE', 'TextRank', 'BERT', 'LSA']
algoritmos_resumenes = [resumenes_tfidf, resumenes_frecuencia, resumenes_rake, 
                        resumenes_textrank, resumenes_bert, resumenes_lsa]

matriz_similitud = np.zeros((6, 6))
for i in range(6):
    for j in range(6):
        if i == j:
            matriz_similitud[i][j] = 100.0
        else:
            matriz_similitud[i][j] = calcular_solapamiento(
                algoritmos_resumenes[i], 
                algoritmos_resumenes[j]
            )

df_similitud = pd.DataFrame(
    matriz_similitud,
    index=algoritmos_nombres,
    columns=algoritmos_nombres
)

print("\nTABLA 3: MATRIZ DE SIMILITUD ENTRE ALGORITMOS (% oraciones comunes)")
print("-" * 80)
print(df_similitud.to_string())
print("=" * 80)

# Visualizaci√≥n de similitud
plt.figure(figsize=(12, 10))
sns.heatmap(df_similitud, annot=True, fmt='.1f', cmap='YlOrRd', 
            square=True, linewidths=1, cbar_kws={'label': '% Similitud'})
plt.title('Figura 5: Mapa de Calor de Similitud entre Algoritmos\n(% de oraciones en com√∫n)', 
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Algoritmo', fontsize=12, fontweight='bold')
plt.ylabel('Algoritmo', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n‚úÖ An√°lisis cualitativo completado")
```

### 13. Evaluaci√≥n Multidimensional y Conclusi√≥n Final 

Esta celda finaliza la pr√°ctica realizando una **evaluaci√≥n hol√≠stica** de los seis algoritmos.
```python
# =============================================================================
# EVALUACI√ìN MULTIDIMENSIONAL
# =============================================================================

# Evaluaci√≥n multidimensional (escala 1-5)
evaluacion = {
    'Algoritmo': ['TF-IDF', 'Frecuencia', 'RAKE', 'TextRank', 'BERT', 'LSA'],
    'Velocidad': [5, 5, 4, 5, 1, 5],
    'Coherencia': [3, 2, 4, 4, 5, 3],
    'Completitud': [4, 3, 4, 4, 5, 4],
    'Preservaci√≥n Formato': [3, 3, 5, 3, 5, 1],
    'Facilidad de Uso': [5, 5, 4, 4, 3, 4]
}

df_eval = pd.DataFrame(evaluacion)
df_eval['Puntuaci√≥n Total'] = df_eval.iloc[:, 1:].sum(axis=1)
df_eval = df_eval.sort_values('Puntuaci√≥n Total', ascending=False)

print("=" * 80)
print("TABLA 4: EVALUACI√ìN MULTIDIMENSIONAL DE ALGORITMOS")
print("=" * 80)
print("Escala: 1 (Bajo) - 5 (Alto)")
print("=" * 80)
print(df_eval.to_string(index=False))
print("=" * 80)

# Visualizaci√≥n radar
fig, axes = plt.subplots(2, 3, figsize=(18, 12), subplot_kw=dict(projection='polar'))
axes = axes.flatten()

# SOLUCI√ìN: Usar los nombres de columnas REALES del DataFrame
categorias_df = ['Velocidad', 'Coherencia', 'Completitud', 'Preservaci√≥n Formato', 'Facilidad de Uso']
# Nombres para mostrar en gr√°fica (con saltos de l√≠nea para mejor visualizaci√≥n)
categorias_display = ['Velocidad', 'Coherencia', 'Completitud', 'Preservaci√≥n\nFormato', 'Facilidad\nde Uso']

num_vars = len(categorias_df)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

colores = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6', '#1abc9c']

for idx, (ax, algoritmo, color) in enumerate(zip(axes, evaluacion['Algoritmo'], colores)):
    # Usar categorias_df (nombres reales de columnas) para extraer valores
    valores = df_eval[df_eval['Algoritmo'] == algoritmo][categorias_df].values.flatten().tolist()
    valores += valores[:1]
    
    ax.plot(angles, valores, 'o-', linewidth=2, color=color, label=algoritmo)
    ax.fill(angles, valores, alpha=0.25, color=color)
    ax.set_xticks(angles[:-1])
    # Usar categorias_display (con saltos de l√≠nea) para etiquetas
    ax.set_xticklabels(categorias_display, size=9)
    ax.set_ylim(0, 5)
    ax.set_yticks([1, 2, 3, 4, 5])
    ax.set_title(f'{algoritmo}\nPuntuaci√≥n: {df_eval[df_eval["Algoritmo"]==algoritmo]["Puntuaci√≥n Total"].values[0]}/25', 
                 size=12, fontweight='bold', pad=20)
    ax.grid(True)

plt.suptitle('Figura 6: Evaluaci√≥n Multidimensional por Algoritmo', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Gr√°fica de barras apiladas
fig, ax = plt.subplots(figsize=(14, 8))

# Usar categorias_df (nombres reales) para extraer datos
df_plot = df_eval.set_index('Algoritmo')[categorias_df]
df_plot.plot(kind='barh', stacked=True, ax=ax, 
             color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6'])

ax.set_title('Figura 7: Comparaci√≥n Acumulativa de Dimensiones por Algoritmo', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xlabel('Puntuaci√≥n Acumulada', fontsize=12, fontweight='bold')
ax.set_ylabel('Algoritmo', fontsize=12, fontweight='bold')
ax.legend(title='Dimensi√≥n', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n‚úÖ Evaluaci√≥n multidimensional completada")
```
